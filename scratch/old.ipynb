# Notebook 03 Гўв‚¬вЂќ CARMA(p, p-1) Kalman quasi-MLE: fixed AR, free MA

Initial values of `b` come from the spectral factorisation (Section 6 of notebook 02, saved in `data/kalman/price_carma_b_init.json`).

**Model:** `dX = A*X dt + ep*sigma dL(t)`,  `Y_t = b^T X_t`

## 1. Data and AR structure

Loads the hourly log-price residuals from `data/seasonality/french_panel.csv` (column `log_price_resid`) and reconstructs the companion matrix `A` from the AR coefficients saved by notebook 02. The AR order `p` is read dynamically from `price_multiscale_carma_selected.json`. Computes the stationary covariance `P_norm` (Lyapunov, sigma=1) and the matrix exponential `eAh` (h=1 hr) Гўв‚¬вЂќ both fixed throughout the optimisation.

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Load hourly series ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
df <- read.csv('../data/seasonality/french_panel.csv', row.names=1)
y  <- as.numeric(df[['log_price_resid']])
y  <- y[is.finite(y)]
y  <- y - mean(y)
n  <- length(y); h <- 1L
cat(sprintf('n = %d,  sd(y) = %.6f\n', n, sd(y)))

# ГўвЂќв‚¬ГўвЂќв‚¬ Load AR structure from multiscale notebook (Section 3) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
ms   <- fromJSON('../data/kalman/price_multiscale_carma_selected.json')
ar   <- as.numeric(ms$ar_coefficients)
p    <- length(ar)
cat(sprintf('AR order p = %d\n', p))

# ГўвЂќв‚¬ГўвЂќв‚¬ Companion matrix A (fixed) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
A_fix <- matrix(0, p, p)
for (i in seq_len(p-1)) A_fix[i, i+1] <- 1.0
A_fix[p, ] <- -rev(ar)

eigs <- sort(Re(eigen(A_fix, only.values=TRUE)$values))
hls  <- log(2) / abs(eigs)
cat('\nEigenvalues (real parts):\n')
for (i in seq_along(eigs))
  cat(sprintf('  lam%d = %+.4e h^-1   hl = %.2f h  (%.1f days)\n',
              i, eigs[i], hls[i], hls[i]/24))

# ГўвЂќв‚¬ГўвЂќв‚¬ Lyapunov: A P + P A^T + ep ep^T = 0  via eigendecomposition ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# Avoids Kronecker form (ill-conditioned for near-zero eigenvalues).
# Formula: P = V * X * V^H,  X_ij = -(V^{-1} Q V^{-H})_ij / (d_i + conj(d_j))
ep     <- rep(0, p); ep[p] <- 1.0
ev     <- eigen(A_fix)
V      <- ev$vectors; d <- ev$values
Vi     <- solve(V)
RHS    <- -(Vi %*% (ep %o% ep) %*% Conj(t(Vi)))
P_norm <- Re(V %*% (RHS / outer(d, Conj(d), "+")) %*% Conj(t(V)))
cat(sprintf('\nP_norm: min_eig=%.3e\n', min(eigen(P_norm, only.values=TRUE)$values)))

# ГўвЂќв‚¬ГўвЂќв‚¬ Matrix exponential e^{A*h}  (h=1 hr, computed once) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
eAh <- expm(A_fix * h)
cat(sprintf('eAh computed (%dx%d).\n', p, p))
````

## 2. Kalman filter functions

Two implementations:

**Stationary Kalman** (fast, used for multi-restart optimization):
`\$\ = P_{\rm norm}b/(b^\top P_{\rm norm}b), \quad \varepsilon_t = y_t - b^\top e^{Ah}\hat x_{t-1},\quad \hat x_t = e^{Ah}\hat x_{t-1} + K\varepsilon_t.\$\$`

Profile NLL (approximate): $\ell_{\rm approx}(b) = \tfrac{n-1}{2}\log SS_{\rm inn}$, $\hat\sigma^2 = SS_{\rm inn}/((n-1)\,b^\top P_{\rm norm}b)$.

**Full Riccati Kalman** (rigorous QMLE, В§19.1, for final $\hat\nu^2$ in Section 4c):
`\$\$\Omega_{i+1} = Q + F\Omega_i F^\top - \Delta_i^{-1}\Theta_i\Theta_i^\top,\quad Q = P_{\rm norm} - F P_{\rm norm} F^\top\$\$`

Profile NLL (exact, eq. 19.5): $\ell = \log(n^{-1}S) + n^{-1}\sum\log\Delta_i$,  $\hat\nu^2 = n^{-1}S = n^{-1}\sum\varepsilon_i^2/\Delta_i$ (eq. 19.4).

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Rcpp: generic p-dimensional stationary Kalman loop ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
cppFunction('
double kalman_ss_cpp(NumericVector y, NumericMatrix eAh,
                     NumericVector b_obs, NumericVector K) {
  int n = y.size(), p = b_obs.size();
  std::vector<double> x(p, 0.0), px(p);
  double ss = 0.0;
  for (int t = 1; t < n; t++) {
    for (int i = 0; i < p; i++) {
      px[i] = 0.0;
      for (int j = 0; j < p; j++) px[i] += eAh(i,j) * x[j];
    }
    double inn = y[t];
    for (int i = 0; i < p; i++) inn -= b_obs[i] * px[i];
    ss += inn * inn;
    for (int i = 0; i < p; i++) x[i] = px[i] + K[i] * inn;
  }
  return ss;
}
')

# ГўвЂќв‚¬ГўвЂќв‚¬ Profile NLL: minimise over b_free = b[2:p], b[1]=1 fixed ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# Correct profile: minimise log(SS_inn); S_norm must NOT appear
# (including -log(S_norm) gives a spurious incentive to inflate b)
carma_nll_profile <- function(b_free, y, P_norm, eAh) {
  b_obs  <- c(1.0, b_free)
  S_norm <- as.numeric(t(b_obs) %*% P_norm %*% b_obs)
  if (!is.finite(S_norm) || S_norm <= 0) return(1e15)
  K_norm <- as.numeric(P_norm %*% b_obs) / S_norm
  ss_inn <- kalman_ss_cpp(y, eAh, b_obs, K_norm)
  if (!is.finite(ss_inn) || ss_inn <= 0) return(1e15)
  0.5 * (length(y) - 1L) * log(ss_inn)
}

# ГўвЂќв‚¬ГўвЂќв‚¬ Recover sigma and full log-likelihood after optimisation ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
carma_recover <- function(b_free, y, P_norm, eAh) {
  b_obs  <- c(1.0, b_free)
  S_norm <- as.numeric(t(b_obs) %*% P_norm %*% b_obs)
  K_norm <- as.numeric(P_norm %*% b_obs) / S_norm
  ss_inn <- kalman_ss_cpp(y, eAh, b_obs, K_norm)
  nm1    <- length(y) - 1L
  sig2   <- ss_inn / (nm1 * S_norm)
  S      <- sig2 * S_norm
  loglik <- -0.5 * (nm1 * log(2*pi) + nm1 * log(S) + ss_inn / S)
  list(b=b_obs, sigma=sqrt(sig2), loglik=loglik)
}

cat('Kalman C++ + NLL functions compiled.\n')
````

## 3. Initialisation and MLE

Initial `b` from the spectral factorisation of the multiscale ACF model (exact ACF representation). Bounds `[b*/3, 3*b*]` (multiplicative, enforce `b_j > 0`). `scale = b_init` in `nlminb` normalises the parameter space to O(1) despite values spanning several orders of magnitude.

Restarts: 15 random starts with multiplicative perturbations `x exp(N(0, 0.3))`, clipped to bounds.

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Load spectral-factorisation init (multiscale notebook, Section 6) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
b_init_data <- fromJSON('../data/kalman/price_carma_b_init.json')
b_full_init <- as.numeric(b_init_data$b_coefficients)
b_free_init <- b_full_init[-1]

cat('=== Starting point from spectral factorisation ===\n')
for (j in seq_along(b_free_init))
  cat(sprintf('  b%d = %.4e\n', j, b_free_init[j]))
cat(sprintf('  sigma_init = %.4e  (method-of-moments, not used in MLE)\n',
            b_init_data$sigma))

# ГўвЂќв‚¬ГўвЂќв‚¬ Bounds: [b*/3, 3*b*] Гўв‚¬вЂќ tight around init (init is already near-optimal)
lower_b <- b_free_init / 3
upper_b <- b_free_init * 3

# Sanity check: NLL and loglik at starting point
nll0 <- carma_nll_profile(b_free_init, y, P_norm, eAh)
rec0 <- carma_recover(b_free_init, y, P_norm, eAh)
cat(sprintf('\nNLL at init  = %.4f\n', nll0))
cat(sprintf('loglik at init = %.4f\n', rec0$loglik))
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Optimisation parameters (edit here) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
N_RESTARTS  <- 1000     # number of random restarts
EVAL_MAX    <- 10000L  # max NLL evaluations per restart
ITER_MAX    <- 10000L  # max nlminb iterations per restart
SD_PERTURB  <- 0.3    # log-space perturbation sd
# ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬

set.seed(42)
best <- NULL

for (i in seq_len(N_RESTARTS)) {
  b0 <- if (i == 1L) b_free_init else
          b_free_init * exp(rnorm(length(b_free_init), 0, SD_PERTURB))
  b0 <- pmax(pmin(b0, upper_b), lower_b)

  res <- tryCatch(
    nlminb(b0, carma_nll_profile,
           y=y, P_norm=P_norm, eAh=eAh,
           lower=lower_b, upper=upper_b,
           scale=b_free_init,
           control=list(eval.max=EVAL_MAX, iter.max=ITER_MAX,
                        rel.tol=1e-10, x.tol=1e-10)),
    error=function(e) list(objective=1e15, par=b0, convergence=-1L)
  )

  rec_i  <- carma_recover(res$par, y, P_norm, eAh)
  cat(sprintf('restart %2d : loglik = %10.4f   conv = %d\n',
              i, rec_i$loglik, res$convergence))
  if (is.null(best) || res$objective < best$objective) best <- res
}
cat(sprintf('\nBest loglik = %.6f\n', carma_recover(best$par, y, P_norm, eAh)$loglik))
````

## 4b. Penalized MLE Гўв‚¬вЂќ ACF penalty

Standard MLE sacrifices the ACF oscillatory structure. Instead we penalise directly on the
weighted ACF residual (same lags and weights $sw_h$ as the multiscale calibration):

$$\ell_{\rm pen}(b) = \ell_{\rm Kalman}(b)
  - \lambda \sum_h sw_h^2\,\bigl(\rho(h;b) - \rho_{\rm emp}(h)\bigr)^2,
\qquad \rho(h;b) = \frac{b^\top e^{Ah} P_{\rm norm} b}{b^\top P_{\rm norm} b}.$$

$\lambda=0$ recovers pure MLE; large $\lambda$ forces good ACF fit.
`LAMBDA` is tunable Гўв‚¬вЂќ start at 500, increase if ACF still degrades.

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ ACF precomputation (same lags/weights as multiscale notebook) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
v0_acf      <- mean(y^2)
lags_acf    <- c(1:336, 384, 432, 504, 600, 720, 960, 1200, 1440, 1800, 2160, 2520)
n_lags      <- length(lags_acf)
rho_emp_acf <- sapply(lags_acf, function(k)
  sum(y[1:(n-k)] * y[(k+1):n]) / ((n-k) * v0_acf))
wts_acf <- 1 / sqrt(lags_acf)
wts_acf[lags_acf <= 72]  <- wts_acf[lags_acf <= 72]  * 2
wts_acf[lags_acf >= 168] <- wts_acf[lags_acf >= 168] * 2
for (a in c(1,6,12,18,24,30,48,72,96,120,144,168,336,720,1440,2160,2520))
  wts_acf[lags_acf == a] <- wts_acf[lags_acf == a] * 8
sw_acf <- sqrt(wts_acf / mean(wts_acf))

cat(sprintf('Computing eAh for %d ACF lags...\n', n_lags))
eAh_lags <- lapply(lags_acf, function(hh) expm(A_fix * hh))
cat('Done.\n')

# ГўвЂќв‚¬ГўвЂќв‚¬ Tunable parameters ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
LAMBDA     <- 10    # ACF penalty weight
N_RESTARTS <- 15
MAXIT      <- 500L
SD_PERTURB <- 0.3
# ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬

carma_nll_pen_acf <- function(b_free, y, P_norm, eAh, eAh_lags,
                               rho_emp_acf, sw_acf, lambda) {
  nll   <- carma_nll_profile(b_free, y, P_norm, eAh)
  if (nll >= 1e14) return(1e15)
  b_obs <- c(1.0, b_free)
  Pb    <- as.numeric(P_norm %*% b_obs)
  var0  <- as.numeric(b_obs %*% Pb)
  if (var0 <= 0) return(1e15)
  rho_mod <- sapply(eAh_lags, function(M) as.numeric(b_obs %*% M %*% Pb) / var0)
  nll + lambda * sum((sw_acf * (rho_mod - rho_emp_acf))^2)
}

set.seed(42)
best_pen <- NULL

for (i in seq_len(N_RESTARTS)) {
  b0 <- if (i == 1L) b_free_init else
          b_free_init * exp(rnorm(length(b_free_init), 0, SD_PERTURB))
  b0 <- pmax(pmin(b0, upper_b), lower_b)

  res <- tryCatch(
    optim(b0, carma_nll_pen_acf,
          y=y, P_norm=P_norm, eAh=eAh,
          eAh_lags=eAh_lags, rho_emp_acf=rho_emp_acf, sw_acf=sw_acf,
          lambda=LAMBDA,
          method='L-BFGS-B',
          lower=lower_b, upper=upper_b,
          control=list(maxit=MAXIT, factr=1e7, pgtol=1e-10)),
    error=function(e) list(value=1e15, par=b0, convergence=-1L)
  )

  rec_i <- carma_recover(res$par, y, P_norm, eAh)
  cat(sprintf('restart %2d : loglik = %10.4f   conv = %d\n',
              i, rec_i$loglik, res$convergence))
  if (is.null(best_pen) || res$value < best_pen$value) best_pen <- res
}

rec_pen   <- carma_recover(best_pen$par, y, P_norm, eAh)
b_opt_pen <- rec_pen$b

cat(sprintf('\n=== ACF-penalized MLE (lambda=%.0f) ===\n', LAMBDA))
cat('  b0 = 1 (fixed)\n')
for (j in 2:p)
  cat(sprintf('  b%d = %.6e   (init: %.6e,  ratio: %.3f)\n',
              j-1L, b_opt_pen[j], b_free_init[j-1L],
              b_opt_pen[j] / b_free_init[j-1L]))
cat(sprintf('  sigma = %.6e\n', rec_pen$sigma))
cat(sprintf('\n  loglik penalized MLE = %.4f\n', rec_pen$loglik))
cat(sprintf('  loglik spectral init = %.4f\n', rec0$loglik))
cat(sprintf('  loglik gain          = %+.4f\n', rec_pen$loglik - rec0$loglik))

out_pen <- '../data/kalman/price_carma_mle_penalized_result.json'
write(jsonlite::toJSON(list(
  b_coefficients  = as.list(rec_pen$b),
  sigma           = rec_pen$sigma,
  loglik          = rec_pen$loglik,
  lambda          = LAMBDA,
  AR_coefficients = as.list(ar),
  AR_order        = p, MA_order = p - 1L,
  note = paste0('CARMA(9,8) ACF-penalized Kalman MLE (lambda=', LAMBDA, ', nb14).')
), auto_unbox=TRUE, digits=15), out_pen)
cat('saved:', out_pen, '\n')
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Final results summary ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
loglik_arma43 <- 55803.6592   # ARMA(4,3) discrete, k=7, fitted in nb03 Section 0
k_carma <- p                  # b1..b_{p-1} free + sigma = p params

cat(sprintf('=== CARMA(%d,%d) ACF-penalized MLE Гўв‚¬вЂќ final parameters ===\n\n', p, p-1))

cat('-- AR coefficients (fixed from multiscale) --\n')
for (j in seq_along(ar))
  cat(sprintf('  a%d = %.6e\n', j, ar[j]))

cat(sprintf('\n-- Eigenvalues of A --\n'))
for (i in seq_along(eigs))
  cat(sprintf('  lam%d = %+.4e h^-1   hl = %.2f h  (%.1f days)\n',
              i, eigs[i], hls[i], hls[i]/24))

cat('\n-- MA coefficients (estimated) --\n')
cat('  b0 = 1 (fixed)\n')
for (j in 2:p)
  cat(sprintf('  b%d = %.6e   (init: %.6e,  ratio: %.3f)\n',
              j-1L, b_opt_pen[j], b_free_init[j-1L],
              b_opt_pen[j]/b_free_init[j-1L]))

cat(sprintf('\n-- Noise --\n'))
cat(sprintf('  sigma = %.6e\n', rec_pen$sigma))

cat(sprintf('\n-- Log-likelihood comparison --\n'))
cat(sprintf('  CARMA(%d,%d) ACF-pen MLE  loglik = %10.4f   k = %d\n',
            p, p-1, rec_pen$loglik, k_carma))
cat(sprintf('  ARMA(4,3)  discrete       loglik = %10.4f   k = 7  (benchmark)\n',
            loglik_arma43))
cat(sprintf('  Gain pen vs init          : %+.2f log-lik units\n',
            rec_pen$loglik - rec0$loglik))

cat(sprintf('\n-- AIC --\n'))
cat(sprintf('  CARMA(%d,%d) pen MLE  AIC = %.2f\n', p, p-1, -2*rec_pen$loglik + 2*k_carma))
cat(sprintf('  ARMA(4,3)            AIC = %.2f\n',           -2*loglik_arma43  + 2*7))
cat(sprintf('  delta-AIC            = %+.2f  (CARMA minus ARMA)\n',
            (-2*rec_pen$loglik + 2*k_carma) - (-2*loglik_arma43 + 2*7)))
````

## 4c. Rigorous QMLE variance вЂ” full Riccati Kalman

The stationary Kalman (Sections 3-4b) fixes $\Omega_i = P_{\rm norm}$ throughout. 
This is **not** the steady-state of the *discrete* Riccati equation:

$$\Omega_{i+1} = Q + F\Omega_i F^\top - \frac{1}{\Delta_i}\Theta_i\Theta_i^\top,$$

with $F = e^{A\Delta}$, $Q = P_{\rm norm} - F P_{\rm norm} F^\top$, $\Theta_i = F\Omega_i b$, $\Delta_i = b^\top\Omega_i b$.

The rigorous QMLE (Brockwell & Davis 2024, eq. 19.4) gives:
$$\hat\nu^2 = \frac{1}{n} S,\qquad S = \sum_{i=1}^n \frac{(y_i-\hat y_i)^2}{\Delta_i}$$

$\hat\nu^2$ is the **definitive variance of the Levy driver** per unit time.

````r
# в”Ђв”Ђ Rigorous nu^2 from full Riccati QMLE (eq. 19.4) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
b_pen_free <- b_opt_pen[-1]    # drop b0=1 (fixed)

rec_qmle <- carma_recover_qmle(b_pen_free, y, P_norm, eAh, Q_disc)

cat('=== QMLE variance - full Riccati Kalman (Brockwell & Davis В§19.1) ===\n\n')
cat(sprintf('  nu^2   = %.6e   (Levy driver variance per hour, eq. 19.4)\n', rec_qmle$nu2))
cat(sprintf('  sigma  = %.6e\n',                                               rec_qmle$sigma))
cat(sprintf('  loglik = %.4f\n',                                               rec_qmle$loglik))

# Compare with old stationary-Kalman sigma
rec_stat <- carma_recover(b_pen_free, y, P_norm, eAh)
cat(sprintf('\n--- Stationary Kalman  sigma = %.6e  (previous approximation)\n', rec_stat$sigma))
cat(sprintf('--- Riccati QMLE       sigma = %.6e  (rigorous)\n',               rec_qmle$sigma))
cat(sprintf('--- ratio Riccati/stationary = %.6f\n', rec_qmle$sigma / rec_stat$sigma))

# Convergence trace of Delta_i over first steps
cat('\n--- Riccati convergence: Delta_i vs S_norm ---\n')
b_tr <- c(1.0, b_pen_free); Om_t <- P_norm; u_t <- rep(0, p)
for (i in 2:25) {
  Ob <- as.numeric(Om_t %*% b_tr)
  D  <- sum(b_tr * Ob)
  Th <- as.numeric(eAh %*% Ob)
  up <- as.numeric(eAh %*% u_t)
  eps <- y[i] - sum(b_tr * up)
  FO  <- eAh %*% Om_t
  Om_t <- Q_disc + FO %*% t(eAh) - outer(Th, Th) / D
  u_t  <- up + Th / D * eps
  if (i <= 6 || i == 25) cat(sprintf('  step %2d: Delta_i = %.6e\n', i, D))
}
cat(sprintf('  S_norm (P_norm ref) = %.6e\n', as.numeric(t(b_tr) %*% P_norm %*% b_tr)))

# в”Ђв”Ђ Save nu2_qmle to JSON в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
pen_json <- fromJSON('../data/kalman/price_carma_mle_penalized_result.json')
pen_json$nu2_qmle    <- rec_qmle$nu2
pen_json$sigma_qmle  <- rec_qmle$sigma
pen_json$loglik_qmle <- rec_qmle$loglik
write(toJSON(pen_json, auto_unbox=TRUE, digits=15),
      '../data/kalman/price_carma_mle_penalized_result.json')
cat('\nSaved nu2_qmle, sigma_qmle to price_carma_mle_penalized_result.json\n')
cat('=> nu2_qmle is the definitive Levy driver variance for downstream use.\n')
````

## 5. LГѓВ©vy increment recovery Гўв‚¬вЂќ BDY (2011) formula

Canonical decomposition of the CARMA(p, p-1) model. For each **real** eigenvalue `lambda_r` of `A`, the driving LГѓВ©vy increments satisfy:

```
dL_n = (1/alpha_r) [ Y^(r)(n) - Y^(r)(n-1) - lambda_r * h/2 * (Y^(r)(n) + Y^(r)(n-1)) ]
```

where `Y^(r)(t) = b(lambda_r) * [R^{-1} X(t)]_r` and `alpha_r = b(lambda_r) / a'(lambda_r)`.

The **slowest real eigenvalue** (least negative, smallest |lambda|) is selected: it minimises the trapezoid discretisation error O(h^2 * lambda^2) while keeping the signal `Y^(r)(n) - Y^(r)(n-1)` detectable.

Parameters: penalized MLE `b` from Section 4b.

````r
# в”Ђв”Ђ Load penalized MLE b + QMLE variance в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
pen  <- fromJSON('../data/kalman/price_carma_mle_penalized_result.json')
b_pen <- as.numeric(pen$b_coefficients)   # length p, b[1]=1

# nu^2_qmle is the rigorous Levy driver variance (full Riccati QMLE, eq. 19.4)
# Falls back to old sigma if not yet computed (run Section 4c first)
nu2_levy <- if (!is.null(pen$nu2_qmle)) pen$nu2_qmle else pen$sigma^2
sig_levy  <- sqrt(nu2_levy)

# в”Ђв”Ђ Stationary Kalman gain (R_obs -> 0) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
S_norm <- as.numeric(t(b_pen) %*% P_norm %*% b_pen)
K_stat <- as.numeric(P_norm %*% b_pen) / S_norm

cat(sprintf('b loaded: p=%d\n', length(b_pen)))
cat(sprintf('nu^2 (QMLE, rigorous) = %.4e\n', nu2_levy))
cat(sprintf('sigma (Levy driver)   = %.4e\n', sig_levy))

# в”Ђв”Ђ Forward Kalman filter в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
X_f <- matrix(0, n, p)
for (i in 2:n) {
  xp       <- as.numeric(eAh %*% X_f[i-1, ])
  inn      <- y[i] - sum(b_pen * xp)
  X_f[i, ] <- xp + K_stat * inn
}
cat(sprintf('Kalman filter done: %d x %d state matrix\n', n, p))
cat(sprintf('X_f range: [%.4e, %.4e]\n', min(X_f), max(X_f)))
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Eigendecomposition of A_fix ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
ev     <- eigen(A_fix)
evals  <- ev$values
R_mat  <- ev$vectors
Rinv   <- solve(R_mat)

# ГўвЂќв‚¬ГўвЂќв‚¬ Real eigenvalues ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
is_real    <- abs(Im(evals)) < 1e-6
real_evals <- sort(Re(evals[is_real]))
cat('Real eigenvalues (sorted):\n')
for (lam in real_evals)
  cat(sprintf('  lambda = %+.6e h^-1   hl = %.1f h  (%.1f days)\n',
              lam, log(2)/abs(lam), log(2)/abs(lam)/24))

# ГўвЂќв‚¬ГўвЂќв‚¬ Select slowest real eigenvalue (least negative Гўв‚¬вЂќ best trapezoid accuracy) ГўвЂќв‚¬
lambda_r <- real_evals[length(real_evals)]
idx_r    <- which(abs(Im(evals)) < 1e-6 &
                  abs(Re(evals) - lambda_r) < 1e-8)[1]
cat(sprintf('\nSelected: lambda_r = %.6e h^-1  (idx=%d)\n', lambda_r, idx_r))

# ГўвЂќв‚¬ГўвЂќв‚¬ b(lambda_r) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
b_poly <- sum(b_pen * lambda_r^(0:(p-1)))   # b(z) = b0 + b1*z + ... + b_{p-1}*z^{p-1}

# ГўвЂќв‚¬ГўвЂќв‚¬ a'(lambda_r) Гўв‚¬вЂќ derivative of a(z) = z^p + a1*z^{p-1} + ... + ap ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
ar_poly <- sum(sapply(0:(p-1), function(k)
  (p - k) * c(1, ar)[k+1] * lambda_r^(p - k - 1)))

alpha_r <- Re(b_poly / ar_poly)
coeff   <- 1 / alpha_r
cat(sprintf('b(lambda_r)  = %.6e\n', Re(b_poly)))
cat(sprintf("a'(lambda_r) = %.6e\n", Re(ar_poly)))
cat(sprintf('1/alpha_r    = %.6e\n', coeff))

# ГўвЂќв‚¬ГўвЂќв‚¬ Canonical state component Y^(r) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
Yr <- Re(b_poly * (Rinv[idx_r, ] %*% t(X_f)))
Yr <- as.numeric(Yr)
cat(sprintf('Y^(r) range: [%.4e, %.4e]\n', min(Yr), max(Yr)))
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Formula (5.3) with trapezoid rule ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# ГЋвЂќL_n = (1/ГЋВ±_r) [Y^(r)(n) - Y^(r)(n-1) - ГЋВ»_r ГѓвЂ” h/2 ГѓвЂ” (Y^(r)(n)+Y^(r)(n-1))]
h_step <- 1.0   # hourly
dL <- numeric(n - 1)
for (i in 2:n) {
  diff_term <- Yr[i] - Yr[i-1]
  int_trap  <- 0.5 * (Yr[i] + Yr[i-1]) * h_step
  dL[i-1]   <- coeff * (diff_term - lambda_r * int_trap)
}
dL <- dL[is.finite(dL)]

cat('=== LГѓВ©vy increments ГЋвЂќL ===\n')
cat(sprintf('  N    = %d\n',   length(dL)))
cat(sprintf('  Mean = %.6e\n', mean(dL)))
cat(sprintf('  SD   = %.6e\n', sd(dL)))
cat(sprintf('  Skew = %.4f\n', mean((dL-mean(dL))^3)/sd(dL)^3))
cat(sprintf('  Kurt = %.4f\n', mean((dL-mean(dL))^4)/sd(dL)^4))
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ NIG utilities ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
log_dnig_fn <- function(x, mu, delta, alpha, beta) {
  gam <- sqrt(alpha^2 - beta^2)
  r   <- sqrt(delta^2 + (x - mu)^2)
  log(alpha * delta / pi) + delta * gam + beta * (x - mu) +
    log(besselK(alpha * r, nu = 1, expon.scaled = TRUE)) - alpha * r - log(r)
}

rinvgauss_fn <- function(n, mu, lam) {
  c2 <- lam / mu; nv <- rnorm(n)^2
  x  <- 1 + nv / (2*c2) - sqrt(nv / c2 + (nv / (2*c2))^2)
  z  <- ifelse(runif(n) <= 1 / (1 + x), x, 1 / x)
  z * mu
}

rnig_fn <- function(n, mu, delta, alpha, beta) {
  gam <- sqrt(alpha^2 - beta^2)
  V   <- rinvgauss_fn(n, mu = delta / gam, lam = delta^2)
  mu + beta * V + sqrt(V) * rnorm(n)
}

# ГўвЂќв‚¬ГўвЂќв‚¬ Student-t (location-scale) utilities ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
log_dt_ls <- function(x, mu, sigma, nu) {
  z <- (x - mu) / sigma
  lgamma((nu + 1) / 2) - lgamma(nu / 2) -
    log(sqrt(nu * pi) * sigma) -
    ((nu + 1) / 2) * log(1 + z^2 / nu)
}

rdt_ls <- function(n, mu, sigma, nu) mu + sigma * rt(n, df = nu)

# ГўвЂќв‚¬ГўвЂќв‚¬ Gaussian ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
mu_g  <- mean(dL); sig_g <- sd(dL)
ll_g  <- sum(dnorm(dL, mu_g, sig_g, log = TRUE)); aic_g <- -2*ll_g + 2*2

# ГўвЂќв‚¬ГўвЂќв‚¬ NIG Гўв‚¬вЂќ NLL ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
nll_nig <- function(par) {
  mu    <- par[1]
  delta <- exp(par[2])
  alpha <- exp(par[3])
  beta  <- tanh(par[4]) * alpha * 0.99
  lls   <- tryCatch(log_dnig_fn(dL, mu, delta, alpha, beta),
                    error = function(e) rep(-1e15, length(dL)))
  if (any(!is.finite(lls))) return(1e15)
  -sum(lls)
}

# Init from moments: alpha/delta from kurtosis, beta from skewness
ku   <- pmax(mean((dL - mu_g)^4) / sig_g^4, 3.01)
sk   <- mean((dL - mu_g)^3) / sig_g^3
a0   <- sqrt(3 / (sig_g^2 * (ku - 3)))
d0   <- sig_g^2 * a0
gam0 <- a0                                         # approx gamma ~ alpha pour beta petit
b0   <- sk * a0 * sqrt(d0 * gam0) / 3             # beta depuis skewness
b0   <- pmax(pmin(b0, a0 * 0.95), -a0 * 0.95)     # clip dans domaine valide
th0  <- c(mu_g, log(d0), log(a0), atanh(b0 / (a0 * 0.99)))

# Restarts: explore positive and negative beta
set.seed(123)
best_nig <- NULL
for (s in seq(-1, 1, by = 0.25)) {
  th_s <- th0 + c(0, 0, 0, s)
  o <- tryCatch(nlminb(th_s, nll_nig,
                       control = list(iter.max = 5000, rel.tol = 1e-12)),
                error = function(e) NULL)
  if (!is.null(o) && is.finite(o$objective) &&
      (is.null(best_nig) || o$objective < best_nig$objective)) best_nig <- o
}

mu_n    <- best_nig$par[1]
delta_n <- exp(best_nig$par[2])
alpha_n <- exp(best_nig$par[3])
beta_n  <- tanh(best_nig$par[4]) * alpha_n * 0.99
ll_nig  <- -best_nig$objective; aic_nig <- -2*ll_nig + 2*4

# ГўвЂќв‚¬ГўвЂќв‚¬ Student-t Гўв‚¬вЂќ NLL ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# par: mu, log(sigma), log(nu - 2)  ->  nu = exp(par[3]) + 2
nll_t <- function(par) {
  mu    <- par[1]
  sigma <- exp(par[2])
  nu    <- exp(par[3]) + 2
  lls   <- tryCatch(log_dt_ls(dL, mu, sigma, nu),
                    error = function(e) rep(-1e15, length(dL)))
  if (any(!is.finite(lls))) return(1e15)
  -sum(lls)
}

nu_init  <- max(4 + 6 / pmax(ku - 3, 0.1), 4.01)
sig_init <- sqrt(sig_g^2 * (nu_init - 2) / nu_init)
th0_t    <- c(mu_g, log(sig_init), log(nu_init - 2))

set.seed(123)
best_t <- NULL
for (s in c(0, 0.5, -0.5, 1, -1)) {
  th_s <- th0_t + c(0, s * 0.3, s * 0.5)
  o <- tryCatch(nlminb(th_s, nll_t,
                       control = list(iter.max = 5000, rel.tol = 1e-12)),
                error = function(e) NULL)
  if (!is.null(o) && is.finite(o$objective) &&
      (is.null(best_t) || o$objective < best_t$objective)) best_t <- o
}

mu_t    <- best_t$par[1]
sigma_t <- exp(best_t$par[2])
nu_t    <- exp(best_t$par[3]) + 2
ll_t    <- -best_t$objective; aic_t <- -2*ll_t + 2*3

# ГўвЂќв‚¬ГўвЂќв‚¬ Summary ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
gam_n <- sqrt(alpha_n^2 - beta_n^2)
sd_nig_th <- sqrt(delta_n * alpha_n^2 / gam_n^3)
sk_nig_th <- 3 * beta_n / (alpha_n * sqrt(delta_n * gam_n))
sd_t_th   <- sigma_t * sqrt(nu_t / (nu_t - 2))

cat(sprintf('Empirical:  mean=%.3e  sd=%.3e  skew=%.4f  kurt=%.2f\n',
            mu_g, sig_g, sk, ku))
cat(sprintf('\nGaussian    loglik=%10.1f  AIC=%12.1f  (k=2)\n', ll_g, aic_g))
cat(sprintf('NIG         loglik=%10.1f  AIC=%12.1f  (k=4)\n', ll_nig, aic_nig))
cat(sprintf('  mu=%.3e  delta=%.3e  alpha=%.3e  beta=%.3e\n',
            mu_n, delta_n, alpha_n, beta_n))
cat(sprintf('  fitted sd=%.3e (emp=%.3e)  fitted skew=%.4f (emp=%.4f)\n',
            sd_nig_th, sig_g, sk_nig_th, sk))
cat(sprintf('Student-t   loglik=%10.1f  AIC=%12.1f  (k=3)\n', ll_t, aic_t))
cat(sprintf('  mu=%.3e  sigma=%.3e  nu=%.4f\n', mu_t, sigma_t, nu_t))
cat(sprintf('  fitted sd=%.3e (emp=%.3e)\n', sd_t_th, sig_g))
cat(sprintf('\nBest model: %s\n',
            c('Gaussian','NIG','Student-t')[which.min(c(aic_g, aic_nig, aic_t))]))

# ГўвЂќв‚¬ГўвЂќв‚¬ Simulation for QQ ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
set.seed(42)
nig_sim <- rnig_fn(length(dL) * 5, mu_n, delta_n, alpha_n, beta_n)
t_sim   <- rdt_ls(length(dL) * 5, mu_t, sigma_t, nu_t)
probs   <- ppoints(length(dL))
q_emp   <- sort(dL)
q_nor   <- qnorm(probs, mu_g, sig_g)
q_nig   <- quantile(nig_sim, probs)
q_t     <- quantile(t_sim,   probs)

# ГўвЂќв‚¬ГўвЂќв‚¬ Plots ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
xr  <- quantile(dL, c(0.0005, 0.9995))
xg  <- seq(xr[1], xr[2], length.out = 2000)
d_g <- dnorm(xg, mu_g, sig_g)
d_n <- exp(log_dnig_fn(xg, mu_n, delta_n, alpha_n, beta_n))
d_t <- exp(log_dt_ls(xg, mu_t, sigma_t, nu_t))

options(repr.plot.width = 15, repr.plot.height = 10)
par(mfrow = c(2, 3), mar = c(4, 4, 3, 1))

# 1. Distribution complete
hist(dL, breaks = 1000, freq = FALSE, col = '#aec6e8', border = '#aec6e8',
     main = 'Full distribution', xlab = 'dL',
     xlim = xr, ylim = c(0, max(d_g, d_n, d_t) * 1.05))
lines(xg, d_g, col = 'red',       lwd = 2)
lines(xg, d_n, col = 'darkgreen', lwd = 2)
lines(xg, d_t, col = 'purple',    lwd = 2)
legend('topright', c('Gaussian','NIG','Student-t'),
       col = c('red','darkgreen','purple'), lwd = 2, bty = 'n', cex = 0.85)

# 2. Left tail
ql  <- quantile(dL, c(0.005, 0.08))
xzl <- seq(ql[1], ql[2], length.out = 1000)
ylim_l <- c(0, max(dnorm(xzl,mu_g,sig_g),
                   exp(log_dnig_fn(xzl,mu_n,delta_n,alpha_n,beta_n)),
                   exp(log_dt_ls(xzl,mu_t,sigma_t,nu_t))) * 1.1)
hist(dL, breaks = seq(min(dL),max(dL),length.out=3000), freq=FALSE,
     col='#aec6e8', border='#aec6e8', main='Left tail (0.5%-8%)',
     xlab='dL', xlim=ql, ylim=ylim_l)
lines(xzl, dnorm(xzl,mu_g,sig_g),                               col='red',       lwd=2)
lines(xzl, exp(log_dnig_fn(xzl,mu_n,delta_n,alpha_n,beta_n)),   col='darkgreen', lwd=2)
lines(xzl, exp(log_dt_ls(xzl,mu_t,sigma_t,nu_t)),               col='purple',    lwd=2)

# 3. Right tail
qr  <- quantile(dL, c(0.92, 0.995))
xzr <- seq(qr[1], qr[2], length.out = 1000)
ylim_r <- c(0, max(dnorm(xzr,mu_g,sig_g),
                   exp(log_dnig_fn(xzr,mu_n,delta_n,alpha_n,beta_n)),
                   exp(log_dt_ls(xzr,mu_t,sigma_t,nu_t))) * 1.1)
hist(dL, breaks = seq(min(dL),max(dL),length.out=3000), freq=FALSE,
     col='#aec6e8', border='#aec6e8', main='Right tail (92%-99.5%)',
     xlab='dL', xlim=qr, ylim=ylim_r)
lines(xzr, dnorm(xzr,mu_g,sig_g),                               col='red',       lwd=2)
lines(xzr, exp(log_dnig_fn(xzr,mu_n,delta_n,alpha_n,beta_n)),   col='darkgreen', lwd=2)
lines(xzr, exp(log_dt_ls(xzr,mu_t,sigma_t,nu_t)),               col='purple',    lwd=2)

# 4. QQ Gaussian
plot(q_nor, q_emp, pch=16, cex=0.25, col='#4472c4',
     main='QQ -- Gaussian', xlab='Theoretical', ylab='Empirical')
abline(0, 1, col='red', lwd=2)

# 5. QQ NIG
plot(q_nig, q_emp, pch=16, cex=0.25, col='#4472c4',
     main='QQ -- NIG', xlab='Theoretical', ylab='Empirical')
abline(0, 1, col='red', lwd=2)

# 6. QQ Student-t
plot(q_t, q_emp, pch=16, cex=0.25, col='#4472c4',
     main=sprintf('QQ -- Student-t (nu=%.2f)', nu_t),
     xlab='Theoretical', ylab='Empirical')
abline(0, 1, col='red', lwd=2)

par(mfrow = c(1, 1))
````

## 6. LГѓВ©vy increment recovery Гўв‚¬вЂќ comparison across real eigenvalues

The CARMA(p, p-1) model has `Kr` real eigenvalues (the remaining are complex-conjugate pairs, which cannot be used directly in the BDY formula). The number of real eigenvalues is determined by the ACF calibration in notebook 02.

In theory all formulae yield the same `L`. In practice they differ due to trapezoid discretisation error Гўв‚¬вЂќ faster eigenvalues (larger |lambda|) incur larger per-step error O(h^2 * lambda^2). Comparing the distributions across eigenvalues is a consistency check: aligned distributions confirm the recovery is stable.

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Function: LГѓВ©vy recovery for a chosen real eigenvalue ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
levy_recover_r <- function(which_real, ev, Rinv, X_f, b_pen, ar, p, h_step = 1.0) {
  evals  <- ev$values
  is_re  <- abs(Im(evals)) < 1e-6
  re_lam <- sort(Re(evals[is_re]))
  lam    <- re_lam[which_real]
  idx    <- which(abs(Im(evals)) < 1e-6 &
                  abs(Re(evals) - lam) < 1e-8)[1]

  b_lam  <- sum(b_pen * lam^(0:(p-1)))
  ap_lam <- sum(sapply(0:(p-1), function(k)
               (p-k) * c(1,ar)[k+1] * lam^(p-k-1)))
  alpha  <- Re(b_lam / ap_lam)
  coeff  <- 1.0 / alpha

  Yr <- Re(b_lam * as.numeric(Rinv[idx, ] %*% t(X_f)))
  nn <- length(Yr)
  dL <- numeric(nn - 1)
  for (i in 2:nn)
    dL[i-1] <- coeff * ((Yr[i] - Yr[i-1]) - lam * 0.5*(Yr[i]+Yr[i-1])*h_step)
  dL <- dL[is.finite(dL)]

  list(dL=dL, lambda_r=lam, alpha_r=alpha, b_lambda=Re(b_lam), ap_lambda=Re(ap_lam), Yr=Yr)
}

# ГўвЂќв‚¬ГўвЂќв‚¬ Eigendecomposition ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
ev_cmp   <- eigen(A_fix)
Rinv_cmp <- solve(ev_cmp$vectors)

# ГўвЂќв‚¬ГўвЂќв‚¬ Run for all real eigenvalues (number determined by ACF calibration) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
n_real <- sum(abs(Im(ev_cmp$values)) < 1e-6)
cat(sprintf('Real eigenvalues available: %d\n', n_real))

r_list <- lapply(seq_len(n_real), function(i)
  levy_recover_r(i, ev_cmp, Rinv_cmp, X_f, b_pen, ar, p))

# ГўвЂќв‚¬ГўвЂќв‚¬ Summary table ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
cat(sprintf('\n=== Levy increment comparison -- %d real eigenvalues ===\n\n', n_real))
cat(sprintf('%-10s  %+12s  %10s  %10s  %9s  %9s  %7s  %7s\n',
            'Eigenval', 'lambda_r', 'b(lam)', "a'(lam)", 'Mean', 'SD', 'Skew', 'Kurt'))
for (i in seq_len(n_real)) {
  r <- r_list[[i]]; dLi <- r$dL
  cat(sprintf('lam%-7d  %+12.4e  %10.4e  %10.4e  %9.2e  %9.2e  %7.3f  %7.1f\n',
              i, r$lambda_r, r$b_lambda, r$ap_lambda,
              mean(dLi), sd(dLi),
              mean((dLi-mean(dLi))^3)/sd(dLi)^3,
              mean((dLi-mean(dLi))^4)/sd(dLi)^4))
}
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Sanity check: drift vs noise per step ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# dX = AX dt + sigma*ep*dL
# Per step h=1: drift contribution ~ A*X_f[n-1,]*h
#               noise contribution ~ sigma * dL * e_p  (only last component)

h_step <- 1.0

# Drift term: A*X per step (n x p matrix)
drift <- X_f[1:(n-1), ] %*% t(A_fix) * h_step   # (n-1) x p

# Noise term: sigma * dL * e_p  (only hits last component = component p)
sigma_mle <- if (!is.null(pen$sigma_qmle)) pen$sigma_qmle else pen$sigma  # QMLE nu^2
noise_ep  <- sigma_mle * dL   # length n-1, enters only component p

# Per-component SD comparison
cat('=== Drift (A*X*h) vs Noise (sigma*dL*ep) Гўв‚¬вЂќ per component ===\n\n')
cat(sprintf('sigma = %.4e\n\n', sigma_mle))
cat(sprintf('%-12s  %12s  %12s  %8s\n',
            'Component', 'SD(drift)', 'SD(noise)', 'ratio'))
for (j in 1:p) {
  sd_drift <- sd(drift[, j])
  sd_noise <- if (j == p) abs(sigma_mle) * sd(dL) else 0.0
  cat(sprintf('X[%d]         %12.4e  %12s  %s\n',
              j, sd_drift,
              if (j == p) sprintf('%.4e', sd_noise) else '     --     ',
              if (j == p) sprintf('%.3f', sd_noise / sd_drift) else ''))
}

cat(sprintf('\n=> noise/drift ratio at component p=%d: %.4f\n',
            p, abs(sigma_mle) * sd(dL) / sd(drift[, p])))

# Global check: total variance of X explained by drift vs Kalman update
innov <- numeric(n - 1)
for (i in 2:n)
  innov[i-1] <- y[i] - sum(b_pen * as.numeric(eAh %*% X_f[i-1, ]))

cat(sprintf('\nSD(innovation)          = %.4e\n', sd(innov)))
cat(sprintf('sigma * sqrt(S_norm)    = %.4e  (expected innovation SD)\n',
            sigma_mle * sqrt(S_norm)))
cat(sprintf('sd(y)                   = %.4e\n', sd(y)))
````

````r
# Noise on X[p]: dL_BDY directly (already = sigma * dL_true)
sd_noise_correct <- sd(dL)   # pas sigma * dL

cat(sprintf('\n=== Drift vs noise at X[p=%d] ===\n', p))
cat(sprintf('SD(drift at X[p])  = %.4e\n', sd(drift[, p])))
cat(sprintf('SD(noise at X[p])  = %.4e  (= SD(dL_BDY))\n', sd_noise_correct))
cat(sprintf('ratio noise/drift  = %.4f\n', sd_noise_correct / sd(drift[, p])))
````

````r
# ГўвЂќв‚¬ГўвЂќв‚¬ Setup ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
sigma_mle <- pen$sigma
dL_true   <- dL / sigma_mle
mu_g      <- mean(dL_true); sig_g <- sd(dL_true)
ku        <- mean((dL_true-mu_g)^4)/sig_g^4
sk        <- mean((dL_true-mu_g)^3)/sig_g^3

cat(sprintf('sigma_mle = %.4e  |  sd=%.4f  skew=%.4f  kurt=%.2f\n',
            sigma_mle, sig_g, sk, ku))

# ГўвЂќв‚¬ГўвЂќв‚¬ Utilities ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
log_dnig_fn <- function(x, mu, delta, alpha, beta) {
  gam <- sqrt(alpha^2 - beta^2)
  r   <- sqrt(delta^2 + (x-mu)^2)
  log(alpha*delta/pi) + delta*gam + beta*(x-mu) +
    log(besselK(alpha*r, nu=1, expon.scaled=TRUE)) - alpha*r - log(r)
}
rinvgauss_fn <- function(n, mu, lam) {
  c2 <- lam/mu; nv <- rnorm(n)^2
  x  <- 1+nv/(2*c2)-sqrt(nv/c2+(nv/(2*c2))^2)
  z  <- ifelse(runif(n)<=1/(1+x),x,1/x); z*mu
}
rnig_fn <- function(n, mu, delta, alpha, beta) {
  gam <- sqrt(alpha^2-beta^2)
  V   <- rinvgauss_fn(n, mu=delta/gam, lam=delta^2)
  mu+beta*V+sqrt(V)*rnorm(n)
}
log_dt_ls <- function(x, mu, sigma, nu) {
  z <- (x-mu)/sigma
  lgamma((nu+1)/2)-lgamma(nu/2)-log(sqrt(nu*pi)*sigma)-((nu+1)/2)*log(1+z^2/nu)
}
rdt_ls <- function(n, mu, sigma, nu) mu + sigma*rt(n, df=nu)

# ГўвЂќв‚¬ГўвЂќв‚¬ Gaussian ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
ll_g  <- sum(dnorm(dL_true, mu_g, sig_g, log=TRUE))
aic_g <- -2*ll_g + 2*2

# ГўвЂќв‚¬ГўвЂќв‚¬ NIG Гўв‚¬вЂќ manual fit ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# Init from moments (scale-invariant: alpha_true = sigma_mle * alpha_BDY)
a0 <- sqrt(3/(sig_g^2*pmax(ku-3,0.01)))   # ~ 0.04
d0 <- sig_g^2 * a0                        # ~ 0.92
b0 <- pmax(pmin(sk*a0*sqrt(d0*a0)/3, a0*0.9), -a0*0.9)
th0_nig <- c(mu_g, log(d0), log(a0), atanh(b0/(a0*0.99)))

nll_nig <- function(par) {
  mu    <- par[1]
  delta <- exp(par[2])
  alpha <- exp(par[3])
  beta  <- tanh(par[4])*alpha*0.99
  lls   <- tryCatch(log_dnig_fn(dL_true, mu, delta, alpha, beta),
                    error=function(e) rep(-1e15, length(dL_true)))
  if (any(!is.finite(lls))) return(1e15)
  -sum(lls)
}

set.seed(42)
best_nig <- NULL
for (s in seq(-1, 1, by=0.25)) {
  th_s <- th0_nig + c(0, 0, 0, s)
  o <- tryCatch(nlminb(th_s, nll_nig,
                       control=list(iter.max=5000, rel.tol=1e-12)),
                error=function(e) NULL)
  if (!is.null(o) && is.finite(o$objective) &&
      (is.null(best_nig) || o$objective < best_nig$objective)) best_nig <- o
}

mu_n <- best_nig$par[1]; delta_n <- exp(best_nig$par[2])
alpha_n <- exp(best_nig$par[3]); beta_n <- tanh(best_nig$par[4])*alpha_n*0.99
ll_nig <- -best_nig$objective; aic_nig <- -2*ll_nig + 2*4
gam_n  <- sqrt(alpha_n^2 - beta_n^2)

# ГўвЂќв‚¬ГўвЂќв‚¬ Student-t Гўв‚¬вЂќ nu constrained > 4 (finite kurtosis) ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
# nu = exp(par[3]) + 4  =>  nu > 4 always
nu_i   <- max(4 + 6/pmax(ku-3, 0.1), 4.1)
th0_t  <- c(mu_g, log(sqrt(sig_g^2*(nu_i-2)/nu_i)), log(nu_i-4))

nll_t <- function(par) {
  mu    <- par[1]
  sigma <- exp(par[2])
  nu    <- exp(par[3]) + 4          # nu > 4 always (finite kurtosis)
  lls   <- tryCatch(log_dt_ls(dL_true, mu, sigma, nu),
                    error=function(e) rep(-1e15, length(dL_true)))
  if (any(!is.finite(lls))) return(1e15)
  -sum(lls)
}

set.seed(123)
best_t <- NULL
for (s in c(0, 0.5, -0.5, 1, -1)) {
  o <- tryCatch(nlminb(th0_t + c(0,s*0.3,s*0.5), nll_t,
                       control=list(iter.max=5000, rel.tol=1e-12)),
                error=function(e) NULL)
  if (!is.null(o) && is.finite(o$objective) &&
      (is.null(best_t) || o$objective < best_t$objective)) best_t <- o
}

mu_t    <- best_t$par[1]
sigma_t <- exp(best_t$par[2])
nu_t    <- exp(best_t$par[3]) + 4
ll_t    <- -best_t$objective; aic_t <- -2*ll_t + 2*3

# ГўвЂќв‚¬ГўвЂќв‚¬ Summary ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
cat(sprintf('\n=== Fits on dL_true = dL_BDY / sigma_mle ===\n'))
cat(sprintf('Gaussian   loglik=%10.1f  AIC=%10.1f\n', ll_g,   aic_g))
cat(sprintf('NIG        loglik=%10.1f  AIC=%10.1f\n', ll_nig, aic_nig))
cat(sprintf('           mu=%.3e  delta=%.3e  alpha=%.4f  beta=%.4f\n',
            mu_n, delta_n, alpha_n, beta_n))
cat(sprintf('           fitted sd=%.4f  fitted skew=%.4f\n',
            sqrt(delta_n*alpha_n^2/gam_n^3),
            3*beta_n/(alpha_n*sqrt(delta_n*gam_n))))
cat(sprintf('Student-t  loglik=%10.1f  AIC=%10.1f  nu=%.4f (>4: finite kurt)\n',
            ll_t, aic_t, nu_t))
cat(sprintf('Best: %s\n',
            c('Gaussian','NIG','Student-t')[which.min(c(aic_g,aic_nig,aic_t))]))

# ГўвЂќв‚¬ГўвЂќв‚¬ Save ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
write(jsonlite::toJSON(list(
  sigma_mle = sigma_mle,
  gaussian  = list(mu=mu_g, sigma=sig_g, loglik=ll_g, AIC=aic_g),
  nig       = list(mu=mu_n, delta=delta_n, alpha=alpha_n, beta=beta_n,
                   loglik=ll_nig, AIC=aic_nig),
  student_t = list(mu=mu_t, sigma=sigma_t, nu=nu_t,
                   loglik=ll_t, AIC=aic_t),
  note='Fitted on dL_true=dL_BDY/sigma_mle. Simulate: X[n]=eAh*X[n-1]+ep*sigma_mle*dL_true_sim'
), auto_unbox=TRUE, digits=15),
'../data/levy/price_carma_levy_params.json')
cat('saved.\n')

# ГўвЂќв‚¬ГўвЂќв‚¬ QQ plots ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
set.seed(42)
nig_sim <- rnig_fn(length(dL_true)*5, mu_n, delta_n, alpha_n, beta_n)
t_sim   <- rdt_ls(length(dL_true)*5, mu_t, sigma_t, nu_t)
probs   <- ppoints(length(dL_true))
q_emp   <- sort(dL_true)
q_nig   <- quantile(nig_sim, probs)
q_t     <- quantile(t_sim,   probs)

xr  <- quantile(dL_true, c(0.001, 0.999))
xg  <- seq(xr[1], xr[2], length.out=1500)
d_g <- dnorm(xg, mu_g, sig_g)
d_n <- exp(log_dnig_fn(xg, mu_n, delta_n, alpha_n, beta_n))
d_t <- exp(log_dt_ls(xg, mu_t, sigma_t, nu_t))

options(repr.plot.width=15, repr.plot.height=7)
par(mfrow=c(1,3), mar=c(4,4,3,1))

hist(dL_true, breaks=200, freq=FALSE, col='#aec6e8', border='#aec6e8',
     main='dL_true distribution', xlab='dL_true', xlim=xr,
     ylim=c(0, max(d_g,d_n,d_t)*1.05))
lines(xg, d_g, col='red',       lwd=2)
lines(xg, d_n, col='darkgreen', lwd=2)
lines(xg, d_t, col='purple',    lwd=2)
legend('topright', c('Gaussian','NIG','Student-t'),
       col=c('red','darkgreen','purple'), lwd=2, bty='n', cex=0.85)

plot(q_nig, q_emp, pch=16, cex=0.25, col='#4472c4',
     main='QQ -- NIG', xlab='Theoretical', ylab='Empirical')
abline(0,1,col='red',lwd=2)

plot(q_t, q_emp, pch=16, cex=0.25, col='#4472c4',
     main=sprintf('QQ -- Student-t (nu=%.2f)', nu_t),
     xlab='Theoretical', ylab='Empirical')
abline(0,1,col='red',lwd=2)
par(mfrow=c(1,1))

# ГўвЂќв‚¬ГўвЂќв‚¬ Simulation with explicit sigma_mle ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬ГўвЂќв‚¬
set.seed(42)
p_ar     <- length(ar)
ep_vec   <- rep(0, p_ar); ep_vec[p_ar] <- 1.0
n_burnin <- 2000

dL_true_sim <- mu_t + sigma_t * rt(n + n_burnin, df=nu_t)

X_cur  <- rep(0, p_ar)
Y_sim  <- numeric(n + n_burnin)
for (i in seq_len(n + n_burnin)) {
  X_cur    <- as.numeric(eAh %*% X_cur) + ep_vec * sigma_mle * dL_true_sim[i]
  Y_sim[i] <- sum(b_pen * X_cur)
}
Y_sim <- Y_sim[(n_burnin+1):(n+n_burnin)]

cat(sprintf('\n=== Simulation (Student-t, nu=%.2f) ===\n', nu_t))
cat(sprintf('SD(y_real) = %.6f\n', sd(y)))
cat(sprintf('SD(y_sim)  = %.6f\n', sd(Y_sim)))
cat(sprintf('ratio      = %.4f\n', sd(Y_sim)/sd(y)))

options(repr.plot.width=15, repr.plot.height=5)
par(mfrow=c(1,2), mar=c(4,4,3,1))
acf(y,     lag.max=168, main='ACF -- Real',      col='#4472c4')
acf(Y_sim, lag.max=168, main='ACF -- Simulated', col='#ed7d31')
par(mfrow=c(1,1))
````

````r
library(expm)

# Load data
df <- read.csv('../data/seasonality/french_panel.csv', row.names=1)
y  <- df$log_price_resid
y  <- y[is.finite(y)]; y <- y - mean(y)
n  <- length(y); v0 <- mean(y^2)

# CARMA(1,0): a(s) = s + a1  ГўвЂ вЂ™  A = [-a1] (1x1)
#             b(s) = 1       ГўвЂ вЂ™  b = [1]   (fixed, q=0)
# Free params: a1 > 0, sigma > 0 (reparametrised as log_a1, log_sigma)

carma10_loglik <- function(theta, y) {
  a1   <- exp(theta[1])
  sig2 <- exp(2 * theta[2])

  # Continuous-time state-space (1x1)
  A     <- matrix(-a1, 1, 1)
  b_obs <- matrix(1,   1, 1)
  ep    <- matrix(1,   1, 1)

  eAh    <- expm(A)                              # matrix exp for h=1h
  K_norm <- matrix(1 / (2*a1), 1, 1)            # Lyapunov: A*K + K*A' + ep*ep' = 0
  Q      <- sig2 * (K_norm - eAh %*% K_norm %*% t(eAh))   # process noise

  # Stationary initialisation
  P     <- sig2 * K_norm
  x_hat <- matrix(0, 1, 1)
  ll    <- 0.0

  for (t in seq_len(n)) {
    x_pred <- eAh %*% x_hat
    P_pred <- eAh %*% P %*% t(eAh) + Q
    e      <- y[t] - t(b_obs) %*% x_pred        # innovation
    F      <- t(b_obs) %*% P_pred %*% b_obs      # innovation variance
    ll     <- ll - 0.5 * (log(2*pi) + log(F[1]) + e[1]^2 / F[1])
    K_g    <- P_pred %*% b_obs / F[1]            # Kalman gain
    x_hat  <- x_pred + K_g * e[1]
    P      <- P_pred - K_g %*% t(b_obs) %*% P_pred
  }
  return(ll)
}

# Optimise
res <- optim(c(log(0.07), log(0.01)),
             function(th) -carma10_loglik(th, y),
             method='L-BFGS-B',
             lower=c(-8, -15), upper=c(4, 2))

a1_mle  <- exp(res$par[1])
sig_mle <- exp(res$par[2])
ll_mle  <- carma10_loglik(res$par, y)
hl_d    <- log(2) / a1_mle / 24

cat(sprintf('=== CARMA(1,0) Kalman MLE ===\n'))
cat(sprintf('  a1     = %.6e  (AR coeff)\n', a1_mle))
cat(sprintf('  b      = [1]   (fixed, q=0)\n'))
cat(sprintf('  sigma  = %.6e\n', sig_mle))
cat(sprintf('  half-life = %.3f d\n', hl_d))
cat(sprintf('  v0_stat   = %.4e  |  empirical v0 = %.4e\n',
            sig_mle^2 / (2*a1_mle), v0))
cat(sprintf('  loglik    = %.4f\n', ll_mle))
cat(sprintf('  AIC       = %.4f   (k=2)\n\n', -2*ll_mle + 4))

# Theoretical ACF: rho(h) = b' * expm(A*h) * K_norm * b / v0_model
A_opt  <- matrix(-a1_mle, 1, 1)
K_norm <- matrix(1/(2*a1_mle), 1, 1)
rho_th <- function(h) as.numeric(exp(-a1_mle * h))   # simplifies to scalar

# Empirical ACF
rho_emp_all <- sapply(1:336, function(k)
  sum(y[1:(n-k)] * y[(k+1):n]) / ((n-k)*v0))
lags_diag <- c(1,6,12,18,24,30,48,72,96,120,144,168,336)
rho_emp_d <- sapply(lags_diag, function(k)
  sum(y[1:(n-k)] * y[(k+1):n]) / ((n-k)*v0))
rmse_ou <- sqrt(mean((rho_th(lags_diag) - rho_emp_d)^2))
cat(sprintf('  RMSE (diag lags) = %.6f\n', rmse_ou))

# Plot
h_seq <- 0:336
options(repr.plot.width=8, repr.plot.height=4)
par(mar=c(4,4,2.5,1))
plot(1:336, rho_emp_all, pch=16, cex=0.4, col='#c0392b',
     xlab='lag (hours)', ylab='autocorrelation',
     main=sprintf('CARMA(1,0): a1=%.2e  hl=%.2fd  loglik=%.1f',
                  a1_mle, hl_d, ll_mle),
     xlim=c(0,336))
lines(h_seq, rho_th(h_seq), lwd=1.8, col='#1f4e79')
abline(v=seq(24,336,24), col='#eeeeee', lwd=0.8)
legend('topright',
       legend=c('empirical ACF',
                sprintf('CARMA(1,0)  RMSE=%.4f', rmse_ou)),
       col=c('#c0392b','#1f4e79'), pch=c(16,NA), lty=c(NA,1),
       pt.cex=0.6, bty='n', cex=0.85)
````
